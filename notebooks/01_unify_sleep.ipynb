```python
import pandas as pd, pathlib

DATA_DIR = pathlib.Path('/content/drive/MyDrive/endoease_data')
df = pd.read_csv(DATA_DIR / 'sleepcycle.csv', sep=';')   # ← semicolon!

# clean columns
df = df.rename(columns={
        'Start': 'start_ts',
        'End': 'end_ts',
        'Sleep quality': 'quality_pct_raw',
        'Time in bed': 'time_in_bed_str'
})

# strip “%” and convert to numeric
df['quality_pct'] = df['quality_pct_raw'].str.rstrip('%').astype(float)

# convert h:mm (or H:MM) string → duration in hours
df['duration_h'] = (
    pd.to_timedelta('0:' + df['time_in_bed_str'])   # prepend 0: to handle “8:32”
      .dt.total_seconds() / 3600
)

# pick the three columns we need now
clean_cols = ['start_ts', 'duration_h', 'quality_pct']
df_clean = df[clean_cols].rename(columns={'start_ts': 'date'})

# save to parquet on Drive
df_clean.to_parquet(DATA_DIR / 'sleep_clean.parquet', index=False)
print(df_clean.head())

```
